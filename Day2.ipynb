```python
import pandas as pd
```

```python
data= {
    "Student": ["S1", "S2", "S3", "S4"],
    "Math": [78, 85, 62, 90],
    "Physics": [80, 88, 70, 95]
}
df = pd.DataFrame(data)
df
```

```python
#Task 1
df["Math"].mean()
```

```python
df["Physics"].mean()
```

```python
#Task 2
df[df["Math"] > 80]
```

```python
#Task 3
df["Total"]= df["Math"] + df["Physics"]
df
```

```python
#Task 4
df[df["Physics"] > 85]
```

```python
#Task 5
df["Average"]= df["Total"]/2
df
```

```python
#Task 6
df.drop(columns=['Math', 'Physics', 'Average'])
```
