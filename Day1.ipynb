*Real-life ML applications

-Machine Learning helps us deciding output of data entry based on previous data entries. It helps us understand complicated data with use of a ML algorithm.

-Supervised Learning: Uses labelled data to get an outcome.

-Unsupervised Learning: Uses unlabelled data to find a label of data.

-Reinforced Learning: Uses feedback based method. When algorithm makes an outcome, you correct it by "teaching" algorithm what the outcome actually is.
