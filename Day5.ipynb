### Mean
* Avg value of the feature
* Data is often centered around mean

### Variance
* Tells how spread out data is from the mean
* Low variance -> Values are close together

### Covariance

Measures how two features move together.
* Positive covariance → increase together
* Negative covariance → one increases, other decreases
* Near zero → little linear relationship

Covariance is the foundation of correlation
